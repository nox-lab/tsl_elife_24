```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

import matplotlib.pyplot as plt
import seaborn as sns

import warnings


sns.set_style("whitegrid")
warnings.filterwarnings('ignore')
```

```python
df_participant = pd.read_csv("output/cs_results/PainValsData_cond.csv")
df_model = pd.read_csv("output/cs_results/PainValsModel_cond.csv")
```

```python
def plot_models_vs_participants_prediction_or_preception(sub_id, df_participant, df_model, condition, c, type_ = 1):
    """
    type_ : int
        1 for perception
        2 for prediction
    """
    df_participant_type = df_participant.loc[(df_participant["PID"] == sub_id) & (df_participant["type"] == type_) & (df_participant["cond"] == condition)]
    df_model_type = df_model.loc[(df_model["PID"] == sub_id) & (df_model["type"] == type_) & (df_model["cond"] == condition) & (df_model["modelName"] == "m2_KF_sliced2_halfStudentT_reparam_confEs_norm")]
    
    participant_data = df_participant_type["value"].tolist()
    model_data = df_model_type["value"].tolist()
    # removing nan values
    cleaned_participant_data, cleaned_model_data = [], []
    for i in range(min(len(participant_data), len(model_data))):
        if (not pd.isna(participant_data[i]) and not pd.isna(model_data[i])):
            cleaned_participant_data.append(participant_data[i])
            cleaned_model_data.append(model_data[i])
            
    all_participant_data.extend(cleaned_participant_data)
    all_model_data.extend(cleaned_model_data)
    # plotting the values with a given color
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    sns.regplot(x=cleaned_participant_data, y=cleaned_model_data, x_ci="ci", ci=0, truncate=False, scatter_kws={'s': 5})
    
    if type_ == 1:
        task = "Perception"
    else:
        task = "Prediction"
    plt.xlabel("Participant " + task)
    plt.ylabel("Model " + task)
```

```python
all_participants = set(df_model["PID"].tolist()).intersection(set(df_participant["PID"].tolist()))

n = len(all_participants)
x = np.random.randn(n)
colors = np.random.rand(len(x),3)
letters_fname = ['A','B','C','D']
l=-1
title_names = ['Vol. high - Stoch. high','Vol. high - Stoch. low','Vol. low - Stoch. high','Vol. low - Stoch. low']
for condition in ["HVHS", "HVLS", "LVHS", "LVLS"]:
# for condition in ["LVLS", "LVHS", "HVHS", "HVLS"]:  
    l+=1
    all_participant_data, all_model_data = [], []
    for i, participant_id in enumerate(all_participants):
        plot_models_vs_participants_prediction_or_preception(participant_id, df_participant, df_model, condition, colors[i], type_ = 1)

    sns.regplot(x=all_participant_data, y=all_model_data, truncate=False, scatter=False, line_kws={"linewidth": 4, "color": "black"})
    corr, _ = pearsonr(all_participant_data, all_model_data)
    text_kwargs = dict(ha='center', va='center', fontsize=12, color='black')
    plt.text(50, 90, r'$r=$' + str(corr)[:4], **text_kwargs)
    # plt.title("participant vs model (perception)")
    plt.title(title_names[l])
    plt.savefig('output/extra_analysis/correlation_beh_model_FigS6/perception_' + condition + '_FigS6'+letters_fname[l]+'.eps', format='eps')
    plt.show()
```

```python
letters_fname = ['E','F','G','H']
l=-1
all_participant_data, all_model_data = [], []
for condition in ["HVHS", "HVLS", "LVHS", "LVLS"]:
# for condition in ["LVLS", "LVHS", "HVHS", "HVLS"]:
    l+=1
    for i, participant_id in enumerate(all_participants):
        plot_models_vs_participants_prediction_or_preception(participant_id, df_participant, df_model, condition, colors[i], type_ = 2)
    sns.regplot(x=all_participant_data, y=all_model_data, truncate=False, scatter=False, line_kws={"linewidth": 4, "color": "black"})
    corr, _ = pearsonr(all_participant_data, all_model_data)
    text_kwargs = dict(ha='center', va='center', fontsize=12, color='black')
    plt.text(50, 90, r'$r=$' + str(corr)[:4], **text_kwargs)
    # plt.title("participant vs model (prediction)")
    plt.title(title_names[l])
    plt.savefig('output/extra_analysis/correlation_beh_model_FigS6/prediction_' + condition + '_FigS6'+letters_fname[l]+'.eps', format='eps')
    plt.show()
```
